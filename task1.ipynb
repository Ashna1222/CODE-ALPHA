```python
import random

def draw_hangman(chances):
    if chances == 6:
        print("________      ")
        print("|      |      ")
        print("|             ")
        print("|             ")
        print("|             ")
        print("|             ")
    elif chances == 5:
        print("________      ")
        print("|      |      ")
        print("|      0      ")
        print("|             ")
        print("|             ")
        print("|             ")
    elif chances == 4:
        print("________      ")
        print("|      |      ")
        print("|      0      ")
        print("|     /       ")
        print("|             ")
        print("|             ")
    elif chances == 3:
        print("________      ")
        print("|      |      ")
        print("|      0      ")
        print("|     /|      ")
        print("|             ")
        print("|             ")
    elif chances == 2:
        print("________      ")
        print("|      |      ")
        print("|      0      ")
        print("|     /|\\     ")
        print("|             ")
        print("|             ")
    elif chances == 1:
        print("________      ")
        print("|      |      ")
        print("|      0      ")
        print("|     /|\\     ")
        print("|     /       ")
        print("|             ")
    elif chances == 0:
        print("________      ")
        print("|      |      ")
        print("|      0      ")
        print("|     /|\\     ")
        print("|     / \\     ")
        print("|             ")

def start_hangman_game():
    chances = 6
    word_list = ["example", "python", "hangman", "challenge", "programming", "dynamic", "random"]
    word = random.choice(word_list)
    temp = "_" * len(word)  # Initialize with underscores without spaces
    found = False

    while True:
        if chances == 0:
            print(f"Sorry! You Lost, the word was: {word}")
            print("Better luck next time")
            break

        print("=== Guess the word ===")
        print(' '.join(temp))  # Add space between underscores for better readability
        print(f"(word has {len(word)} letters)")
        print(f"Chances left: {chances}")
        character = input("Enter the character you think the word may have: ")

        if len(character) > 1 or not character.isalpha():
            print("Please enter a single alphabet only")
            continue
        else:
            for num, char in enumerate(list(word)):
                if char == character:
                    temp = temp[:num] + char + temp[num+1:]  # Update the temp string directly
                    found = True

            if found:
                found = False
            else:
                chances -= 1

            if '_' not in temp:
                print(f"\nYou Won! The word was: {word}")
                print(f"You got it in {7 - chances} guesses!")
                break
            else:
                draw_hangman(chances)

            print()

print("===== Welcome to the Hangman Game =====")

while True:
    choice = input("Do you wanna play hangman? (yes/no): ")

    if 'yes' in choice.lower():
        start_hangman_game()
    elif 'no' in choice.lower():
        print('Quitting the game...')
        break
    else:
        print("Please enter a valid choice.")

    print("\n")
```
